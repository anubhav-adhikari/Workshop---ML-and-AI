```python
#Q. no. 7

import matplotlib.pyplot as plt

# Provided temperature data
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

### Task 1: Classify Temperatures ###
cold = []        # Temperatures below 10Â°C
mild = []        # Temperatures between 10Â°C and 15Â°C
comfortable = [] # Temperatures between 15Â°C and 20Â°C

# Classify temperatures into categories
for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp < 15:
        mild.append(temp)
    elif 15 <= temp <= 20:
        comfortable.append(temp)

# Print the classified lists
print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)

### Task 2: Count occurrences in each category ###
print("\nTask 2 Results:")
print(f"Number of cold readings: {len(cold)}")
print(f"Number of mild readings: {len(mild)}")
print(f"Number of comfortable readings: {len(comfortable)}")

### Task 3: Convert to Fahrenheit ###
temperatures_fahrenheit = [(temp * 9/5) + 32 for temp in temperatures]
print("\nTemperatures in Fahrenheit:")
print(temperatures_fahrenheit)

### Task 4: Analyze Temperature Patterns by Time of Day ###
# Separate readings into night (00-08), evening (08-16), and day (16-24)
night = []
evening = []
day = []

for i in range(len(temperatures)):
    if i % 3 == 0:       # Night (00-08)
        night.append(temperatures[i])
    elif i % 3 == 1:     # Evening (08-16)
        evening.append(temperatures[i])
    else:                # Day (16-24)
        day.append(temperatures[i])

# Calculate and print average day-time temperature
average_day_temp = sum(day) / len(day)
print("\nTask 4 Results:")
print(f"Average daytime temperature: {average_day_temp:.2f}Â°C")

# Optional: Plot "day vs. temperature"
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(day) + 1), day, marker='o', linestyle='-', color='orange', label='Day Temperatures')
plt.title("Day vs. Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q. no. 8

#Task 1

def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a nested list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

# Test cases
nested_list_1 = [1, [2, [3, 4], 5], 6, [7, 8]]
nested_list_2 = [1, [2, 3], [4, [5]]]

print(f"Sum of {nested_list_1}: {sum_nested_list(nested_list_1)}")  # Output: 36
print(f"Sum of {nested_list_2}: {sum_nested_list(nested_list_2)}")  # Output: 15

#Task 2

def generate_permutations(s):
    """
    Generate all unique permutations of the given string.
    """
    if len(s) == 1:
        return [s]
    
    permutations = []
    for i, char in enumerate(s):
        # Exclude the current character and get permutations of the remaining string
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            new_perm = char + perm
            if new_perm not in permutations:  # Avoid duplicates
                permutations.append(new_perm)
    return permutations

# Test cases
print(f"Permutations of 'abc': {generate_permutations('abc')}")  # ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
print(f"Permutations of 'aab': {generate_permutations('aab')}")  # ['aab', 'aba', 'baa']

#Task 3

def calculate_directory_size(directory):
    """
    Calculate the total size of a directory, including nested directories and files.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # Check if the value is a nested directory
            total_size += calculate_directory_size(value)  # Recursively sum sizes
        else:
            total_size += value  # Add file size
    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test case
print(f"Total directory size: {calculate_directory_size(directory_structure)} KB")  # Output: 1400 KB
```

```python
# Task 1: Minimum Coins
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    Parameters:
    coins (list of int): A list of coin denominations available for making change.
    amount (int): The target amount for which we need to find the minimum number of coins.
    Returns:
    int: The minimum number of coins required to make the given amount.
         If it is not possible to make the amount with the given coins, returns -1.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Example usage for Task 1
print(min_coins([1, 2, 5], 11))  # Output: 3


# Task 2: Longest Common Subsequence
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence between two strings.
    Parameters:
    s1 (str): The first string.
    s2 (str): The second string.
    Returns:
    int: The length of the longest common subsequence.
    """
    length1 = len(s1)
    length2 = len(s2)
    dp = [[0] * (length2 + 1) for _ in range(length1 + 1)]  # Fixed syntax

    for i in range(1, length1 + 1):
        for j in range(1, length2 + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[length1][length2]

# Example usage for Task 2
print(longest_common_subsequence("abcde", "ace"))  # Output: 3


# Task 3: Knapsack Problem
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming.
    Parameters:
    weights (list of int): The weights of the items.
    values (list of int): The values of the items.
    capacity (int): The maximum weight capacity of the knapsack.
    Returns:
    int: The maximum value that can be obtained within the given weight capacity.
    """
    dp = [[0] * (capacity + 1) for _ in range(len(weights) + 1)]

    for i in range(1, len(weights) + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], values[i - 1] + dp[i - 1][w - weights[i - 1]])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[len(weights)][capacity]

# Example usage for Task 3
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))  # Output: 9
```
