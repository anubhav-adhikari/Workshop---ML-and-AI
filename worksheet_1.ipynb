```python
import numpy as np

# Arrays A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Task 1: Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("A * A^-1 =\n", identity)
print("Verification: Is A * A^-1 equal to I?", np.allclose(identity, np.eye(2)))

# Task 2: Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB =\n", AB)
print("BA =\n", BA)
print("Verification: Is AB equal to BA?", np.array_equal(AB, BA))

# Task 3: Prove (AB)^T = B^T A^T
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)^T =\n", AB_transpose)
print("B^T * A^T =\n", BT_AT)
print("Verification: Is (AB)^T equal to B^T A^T?", np.allclose(AB_transpose, BT_AT))

# Task 4: Solve system of linear equations using inverse methods
# Representing equations in matrix form: AX = B
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Solve for X using np.linalg.inv
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("\nSolution to the linear equations using inverse method:")
print("X =", X)

# Alternative: Solve directly using np.linalg.solve
X_direct = np.linalg.solve(A_eq, B_eq)
print("\nSolution to the linear equations using np.linalg.solve:")
print("X =", X_direct)
```

```python
import numpy as np
import time

# Task 1: Element-wise Addition
size = 1_000_000
list1 = [i for i in range(size)]
list2 = [i for i in range(size)]
array1 = np.arange(size)
array2 = np.arange(size)

# Python Lists
start_time = time.time()
list_addition = [list1[i] + list2[i] for i in range(size)]
time_list_add = time.time() - start_time

# NumPy Arrays
start_time = time.time()
array_addition = array1 + array2
time_array_add = time.time() - start_time

print(f"Element-wise addition:\nPython Lists: {time_list_add:.6f} seconds\nNumPy Arrays: {time_array_add:.6f} seconds")

# Task 2: Element-wise Multiplication
# Python Lists
start_time = time.time()
list_multiplication = [list1[i] * list2[i] for i in range(size)]
time_list_mul = time.time() - start_time

# NumPy Arrays
start_time = time.time()
array_multiplication = array1 * array2
time_array_mul = time.time() - start_time

print(f"\nElement-wise multiplication:\nPython Lists: {time_list_mul:.6f} seconds\nNumPy Arrays: {time_array_mul:.6f} seconds")

# Task 3: Dot Product
# Python Lists
start_time = time.time()
dot_product_list = sum(list1[i] * list2[i] for i in range(size))
time_list_dot = time.time() - start_time

# NumPy Arrays
start_time = time.time()
dot_product_array = np.dot(array1, array2)
time_array_dot = time.time() - start_time

print(f"\nDot product:\nPython Lists: {time_list_dot:.6f} seconds\nNumPy Arrays: {time_array_dot:.6f} seconds")

# Task 4: Matrix Multiplication
matrix_size = 1000
list_matrix1 = [[j for j in range(matrix_size)] for i in range(matrix_size)]
list_matrix2 = [[j for j in range(matrix_size)] for i in range(matrix_size)]
array_matrix1 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
array_matrix2 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)

# Python Lists
start_time = time.time()
list_matrix_mult = [
    [sum(list_matrix1[i][k] * list_matrix2[k][j] for k in range(matrix_size)) for j in range(matrix_size)]
    for i in range(matrix_size)
]
time_list_matrix_mult = time.time() - start_time

# NumPy Arrays
start_time = time.time()
array_matrix_mult = np.dot(array_matrix1, array_matrix2)
time_array_matrix_mult = time.time() - start_time

print(f"\nMatrix multiplication:\nPython Lists: {time_list_matrix_mult:.6f} seconds\nNumPy Arrays: {time_array_matrix_mult:.6f} seconds")
```
